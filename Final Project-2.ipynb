```python
import random

class RockPaperScissors:
    #Class to handle an instance of a Rock-Paper-Scissors game with unlimited rounds.

    def __init__(self): #Initialize the variables for the class
        self.ties = 0
        self.losses = 0
        self.wins = 0
    
    def usermove(self):
        user_choice = input("Choose: Rock, Paper, or Scissors") #lets user choose what move to play
        if user_choice == "Rock":
            user_move = "Rock" #checks if the user chose rock, and assigns it
            print("You chose",user_move)
            return user_move
        elif user_choice == "Paper":
            user_move = "Paper"
            print("You chose",user_move)
            return user_move
        elif user_choice == "Scissors":
            user_move = "Scissors"
            print("You chose",user_move)
            return user_move
        else:
            user_move = "Error" #lets user know they did the input wrong
            print("Not a valid option. Did you remember to capitalize?")
            return user_move

    def randomcomp(self):
        moves = ["Rock", "Paper", "Scissors"]
        comp_move = random.choice(moves) #randomly chooses one of the 3 moves
        return comp_move

    def results(self, user, computer):
        print("Rock...")
        print("Paper...")
        print("Scissors!")
        print("Computer chose", computer)
        if user == "Rock": #Outputs the result if user chose rock
            if computer == "Rock":
                print("It's a tie!")
                self.ties += 1
            elif computer == "Paper":
                print("You lose!")
                self.losses += 1
            elif computer == "Scissors":
                print("You win!")
                self.wins += 1
            else:
                print("There was an error, and the game could not be played.")
        elif user == "Paper":
            if computer == "Rock":
                print("You win!")
                self.wins += 1
            elif computer == "Paper":
                print("It's a tie!")
                self.ties += 1
            elif computer == "Scissors":
                print("You lose!")
                self.losses += 1
            else:
                print("There was an error, and the game could not be played.")
        elif user == "Scissors":
            if computer == "Rock":
                print("You lose!")
                self.losses += 1
            elif computer == "Paper":
                print("You win!")
                self.wins += 1
            elif computer == "Scissors":
                print("It's a tie!")
                self.ties += 1
            else:
                print("There was an error, and the game could not be played.")
        else: 
            print ("There was an error, and the game could not be played.") #outputs if there was something wrong with user's input
        
    def scoreboard(self):
        print ("You have", self.wins, "wins,", self.losses, "losses,", "and", self.ties, "ties")   
        
        
if __name__ == "__main__": # Initialize an instance of RockPaperScissors for people to refer to
    game = RockPaperScissors() # Keep playing the came repeatedly, stop playing by exiting
    # the entire program directly.
    while True:
        user = game.usermove()
        computer = game.randomcomp()
        game.results(user, computer)
        game.scoreboard()# Run a round of RPS

        # Find out if the user want to continue playing or not.
        while True:
            continue_prompt = input('\nDo you wish to play again? (y/n): ').lower()
            if continue_prompt == 'n':
                # Exit the game directly after printing a response.
                print("Bye, see you next time!")
                exit()
            elif continue_prompt == 'y':
                # Break the continue prompt loop and keep playing.
                break
            else:
                # Bad input was given, re-request if we want to play again.
                print("Invalid input!\n")
                continue
```
